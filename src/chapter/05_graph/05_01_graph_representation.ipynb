# Graph representation

The following examples show how graphs can be internally represented.

All examples utilize the `graph` module in the corresponding subfolder.

```python
from importlib import reload

import logging
logging.basicConfig(level=logging.WARN,format='%(levelname)s - %(message)s')
logging.getLogger("graph.undirectedgraph").setLevel(logging.INFO)
```

Visualize a graph with nodes {A, B, C, D, E} and some edges between those nodes.

```python
%matplotlib inline

# Prevent caching of modules
from importlib import reload

import networkx as nx

from graph.undirectedgraph import * 
import matplotlib.pyplot as plt

g = Graph()
g.add_node(Node("A"))
g.add_node(Node("B"))
g.add_node(Node("C"))
g.add_node(Node("D"))
g.add_node(Node("E"))

g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(1, 4)

nx.draw(g.toNetworkx(), with_labels=True, font_weight="bold", node_size=500, alpha=0.3)
plt.show()
```

Print the internal adjacency matrix representing this graph.

```python
# Adjacency matrix
g.print_matrix()
```

Alternatively, a graph can be stored based on "adjacency lists".
The Graph class can return them, with the label or internal index - or the node as index, whatever is needed.

```python
print(f"Graph Adjacency List (label): {g.getLabelGraphAdjacencyList()}")
print(f"Graph Adjacency List (index): {g.getIndexGraphAdjacencyList()}")
print(f"Graph Adjacency List (nodes): {g.getNodeGraphAdjacencyList()}")
```
