```python
%matplotlib inline
import lzw.lzw
```

Encode / Decode using the LZW algorithm.

```python
#data = "LZWLZ78LZ77LZCLZM"
data = "LZWLZ78LZ77LZCLZMWLZAP"
#data = "TOBEORNOTTOBEORTOBEORNOT"
#data = "ALOLLALELALEALOHELOLLALLELLALELOLAALELL"

encoded_data, _ = lzw.lzw.lzw_compress(data)
decoded_data = lzw.lzw.lzw_decompress(encoded_data)

print(f"data:         {data}")
print(f"encoded_data: {encoded_data}")
print(f"size(data): {len(data)}*8={len(data)*8} | size(encoded): {len(encoded_data)}*12={len(encoded_data)*12} ")
print(f"decoded_data: {decoded_data}")
```

```python
data = "TOBEORNOTTOBETOBEORNOTTOBE"

encoded_data, dictionary = lzw.lzw.lzw_compress(data, initializeDictionaryFromASCII=False)
decoded_data = lzw.lzw.lzw_decompress(encoded_data, initializeDictionaryFromASCII=False, decode_dictionary=dictionary)

print(f"data:         {data}")
print(f"encoded_data: {encoded_data}")
print(f"dictionary  : {dictionary}")
print(f"decoded_data: {decoded_data}")
```
